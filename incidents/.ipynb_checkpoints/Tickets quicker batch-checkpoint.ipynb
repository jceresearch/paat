```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(19680801)
import pandas as pd
from faker import Faker
faker = Faker()
from datetime import datetime 
from datetime import timedelta
import matplotlib.pyplot as plt
def simulate_duration(datetime_created):
    p= np.random.uniform()
    duration= 90*p*24*60
    end_date= datetime_created + timedelta(minutes= duration)
    if end_date >= datetime.now():
        return -1
    return duration
def simulate_team():
    t=np.random.choice([1,2,3,4,5], 1, p=[0.5, 0.1, 0.1, 0.25,.05])[0]
    return "Team "+str(t)
def simulate_priority():
    p=np.random.choice([1,2,3,4], 1, p=[0.001, 0.01, 0.3, 0.689])[0]
    return p
        

row_list=[]
for n in range(0,2000):
    dict1={}
    dict1["created_on"]=faker.date_time_between(start_date='-90d', end_date='now')
    dict1['team']= simulate_team()
    dict1['priority']=simulate_priority()
    dict1["duration"]= simulate_duration(dict1["created_on"])   
    if dict1["duration"]> -1:
        dict1["closed_on"]= dict1["created_on"]+ timedelta(minutes= dict1['duration'])
        dict1['status']='Closed'
    else:
        dict1["closed_on"]=None
        dict1['status']='Open'
        duration_to_date=datetime.now()-dict1['created_on']
        dict1['duration']=duration_to_date.seconds/60
    dict1["duration_days"]= dict1["duration"]/(60*24)
    row_list.append(dict1)
df=pd.DataFrame(row_list)
#df.set_index("created_on", inplace=True)
df.insert(0, 'id', range(880, 880 + len(df)))

df['week_commencing_date'] = pd.to_datetime(df['created_on']) - pd.to_timedelta(7, unit='d')
df['wc']= df['week_commencing_date'].dt.floor('d')
df['created_day']=df['created_on'].dt.floor('d')
df['closed_day']=df['closed_on'].dt.floor('d')
df.set_index('id')
df
```

```python
df2 = df.groupby(['status','team', pd.Grouper(key='wc', freq='W-MON')])['id'].count().reset_index().sort_values('wc')
#pd.pivot_table(df2, index=['wc',"team"], values=["id"], aggfunc=np.sum, columns=["status"], fill_value=0).rename(columns={'id': 'count'})
#use len for count

team_list= df.team.unique()
priority_list= df.priority.unique()
```

```python
def daterange(start_date, end_date):
    for n in range(int ((end_date - start_date).days)):
        yield start_date + timedelta(n)

def flag_df(df):
    if (df.flag_in==0) & (df.flag_out==0):
        return 1
    else:
        return 0

def calculate_queue(d,df):    
    dfq= df[(df['created_day']<=d) & (pd.isnull(df['closed_day'])  | (df['closed_day']>=d))].copy()
    dfq['date']=d
    dfq['flag_in'] = dfq.apply(lambda row: 1 if row['created_day'] ==d  else 0,axis=1)
    dfq['flag_out'] = dfq.apply(lambda row: 1 if row['closed_day'] ==d  else 0,axis=1)
    dfq['flag_wip']=0
    dfq['flag_wip']= dfq.apply(flag_df, axis = 1)
    dfqg=dfq.groupby(['date','priority','team']).agg({'id': len, 'flag_in': np.sum, 'flag_out': np.sum, 'flag_wip':np.sum})  
    return dfqg

d1=datetime(2020,3,1)
d2=datetime(2020,3,31)
d=datetime(2020,5,5)
listofdf=[]



for d in daterange(d1, d2):
   listofdf.append(calculate_queue(d,df).reset_index())
```

```python
dfconso= pd.DataFrame()

for n in range(len(listofdf)):
    dfday=listofdf[n]
    dfconso= dfconso.append(dfday, ignore_index=True) 
```
