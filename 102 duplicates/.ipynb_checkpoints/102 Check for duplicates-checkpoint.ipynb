```python

df.fillna("", inplace=True)
df.drop(df.columns[0],axis=1,inplace=True)
```

```python
df_duplicates = df[df.duplicated(['customer_id'],keep=False)]
df_duplicates

```

```python
df_duplicates.to_excel("./output/102_Duplicates.xlsx", sheet_name='Duplicates')
```
