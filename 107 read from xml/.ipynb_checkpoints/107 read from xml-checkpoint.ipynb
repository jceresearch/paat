```python
import xml.etree.ElementTree as et 
import pandas as pd
```

```python

xtree = et.parse("./test_data/food hygiene rating city of london.xml")
root = xtree.getroot()
df_cols = ["BusinessName", "AddressLine1", "AddressLine2","RatingValue" ]
```

```python
rows = []
for node in root.findall("./EstablishmentCollection/EstablishmentDetail"):
    s_business_name = "" if node.find("BusinessName") is None else  node.find("BusinessName").text
    s_address_line1 = "" if node.find("AddressLine1") is None else  node.find("AddressLine1").text
    s_address_line2 = "" if node.find("AddressLine2") is None else  node.find("AddressLine2").text
    s_rating = "" if node.find("RatingValue") is None else  node.find("RatingValue").text

    rows.append( [s_business_name, s_address_line1, s_address_line2, s_rating])

df = pd.DataFrame(rows, columns = df_cols)
```
