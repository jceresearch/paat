```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']},
                    index=[0, 1, 2, 3])
   

df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                   'B': ['B4', 'B5', 'B6', 'B7'],
                   'C': ['C4', 'C5', 'C6', 'C7'],
                   'D': ['D4', 'D5', 'D6', 'D7']},
                    index=[4, 5, 6, 7])
  
df3 = pd.DataFrame({'A': ['A8', 'A8', 'A10', 'A11',"A12"],
                    'B': ['B8', 'B8', 'B10', 'B11','B12'],
                    'C': ['C8', 'C8', 'C10', 'C11','C12'],
                    'D': ['D8', 'D8', 'D10', 'D11','D12']},
                    index=[8, 9, 10, 10, 12])
df4 = pd.DataFrame({'B': ['B2', 'B3', 'B6', 'B7'],
                    'D': ['D2', 'D3', 'D6', 'D7'],
                    'F': ['F2', 'F3', 'F6', 'F7']},
                     index=[2, 3, 6, 7])
                     
df5 = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                    'B': ['B0', 'B1', 'B2'],
                    'C': ['C0', 'C1', 'C2'],
                    'E': ['E0', 'E1', 'E2']},
                    index=[0, 1, 2])             
df6 = pd.DataFrame({'A': ['A0', 'A1', 'A4'],
                    'B': ['B0', 'B1', 'B4'],
                    'C': ['C0', 'C1', 'C4'],
                    'E': ['E0', 'E1', 'E4']},
                    index=[0, 1, 4])       

df7 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3','A0','A0','A10'],
                    'G': ['G0', 'G1', 'G2', 'G3','G4','G5','G6'],
                    'H':  [1,2,3,4,5,6,7]
                   })
                    
```

Now we concatenate dataframes, which is stitching them toghether either stacked or alongside.
Note that we can have any number of duplications or inconsistencies, the index at this point do not enforce uniqueness, this is not a database system.  There are ways of adding unique constraints but ultimately you need to build it up manually. The best bet is to ensure that kind of controls from a well developed and tested database source and run a few controls to check.

```python
frames = [df1, df2, df3]
pd.concat(frames)
```

Concat() can be used to join two dataframes, but the recommended approach is to do it using Merge()  
Use Concat() if the data is very homogeneous, eg if the report has been split in exact chunks


```python
pd.concat([df1, df4], axis=1, sort=False) # This is joining using the index 
```

Using Merge() is similar to the JOIN command in SQL and offers more options

https://pandas.pydata.org/pandas-docs/stable/user_guide/merging.html

```python
pd.merge(df1, df4, on='B', how='outer') # how can be outer, left , right, inner
```
