```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import pandas as pd


import numpy as np

import seaborn

import matplotlib.pyplot as plt

%matplotlib inline
```

Creating Synthetic data

here is how you create a data set with 200 samples that has 2 features and 4 cluster centers. The standard deviation within each cluster will be set to 1.8.

Then we plot all of the rows in the first column of our data set against all of the rows in the second column of our data set:

```python
raw_data = make_blobs(n_samples = 200, n_features = 2, centers = 4, cluster_std = 1.8)
plt.scatter(raw_data[0][:,0], raw_data[0][:,1], c=raw_data[1])
```

```python
model = KMeans(n_clusters=4)
```

```python
model.fit(raw_data[0])
```

```python
model.labels_
```

```python
model.cluster_centers_
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(10,6))

ax1.set_title('Our Model')

ax1.scatter(raw_data[0][:,0], raw_data[0][:,1],c=model.labels_)

ax2.set_title('Original Data')

ax2.scatter(raw_data[0][:,0], raw_data[0][:,1],c=raw_data[1])
```
