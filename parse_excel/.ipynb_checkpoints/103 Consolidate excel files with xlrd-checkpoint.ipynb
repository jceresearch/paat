```python
#!Python3
import pandas as pd
import os 
from time import gmtime, strftime
import datetime
import xlrd  
import string
import glob

#These are hardcoded values to check
anchor_value="Customer_ID"
anchor_c=0
anchor_r=0
batchfolder= "testdata/batch/"
outputfolder="output/"
```

```python
# Aux function to consolidate many  data frame

def read_workbook(book, anchor_value, anchor_c, anchor_r):
    #We receive a book object already opened by xlrd and return a list of dataframes
    #We also add a timestamp
    sheet_frames=[] 
    ts= strftime("%Y-%m-%d %H:%M:%S")
    n = book.nsheets
    for x in range(0, n):
        sh=book.sheet_by_index(x)
        cellAnchor= sh.cell_value(anchor_r,anchor_c)
        if cellAnchor==anchor_value:  
            print ("Processing sheet " + book.sheet_names()[x])
            num_cols = sh.ncols   # Number of columns
            num_rows = sh.nrows
            print (num_cols, num_rows)
            data=[]
            for row_idx in range(0, num_rows):    # Iterate through rows use sh.nrows for all rows    
                row_data=[]
                for col_idx in range(0, num_cols):  # Iterate through columns

                    cell_obj = sh.cell(row_idx, col_idx)  # Get cell object by row, col
                    cell_value= cell_obj.value
                    cell_type = sh.cell_type(row_idx, col_idx)

                    if cell_type == xlrd.XL_CELL_DATE:
                        get_col= xlrd.xldate.xldate_as_datetime (cell_value, book.datemode)
                    elif cell_type == xlrd.XL_CELL_NUMBER:
                        get_col = int(cell_value)
                    else:
                        v= str(cell_value)
                        stripped= lambda s: "".join(i for i in s if 31 < ord(i) < 127)
                        get_col = stripped(v)

                    row_data.append(get_col)
                data.append(row_data)

            #We create a frame with the first row beign the column names
            frame=pd.DataFrame(data[1:],columns=data[0])
            #We add here a column for the source sheet and a timestamp
            frame["SourceSheet"]=sh.name
            frame["Timestamp"]= ts
            sheet_frames.append (frame)
            print(list(frame))
        else:
            print("Anchor value is not " + anchor_value + ", skipping sheet "+ book.sheet_names()[x])
    return sheet_frames
```

```python
#Aux functions to count csv lines to check completness
def bufcount(filename):
    #this uses a buffer 
    f = open(filename)                  
    lines = 0
    buf_size = 1024 * 1024
    read_f = f.read # loop optimization

    buf = read_f(buf_size)
    while buf:
        lines += buf.count('\n')
        buf = read_f(buf_size)
    return lines
def simplecount(filename):
    lines = 0
    for line in open(filename):
        lines += 1
    return lines
```

```python

#Use one of these to check if we saved the CSV correctly
consofullpath= outputfolder+"conso.csv"
print("Total lines in the file, includes any header line", bufcount(consofullpath))
#print (simplecount(consofullpath))

#Use this one to check if it loads properly, could be issues with split line or some other noise in the data
df2 = pd.read_csv(consofullpath, sep=';', quotechar='"')
print("Total lines loaded into the DF", len(df2))

#This is a check for potentially broken CSV lines as they may have unusual number of separators
f= open(outputfolder+"exceptions.txt","w+")
split_lines = 0
n=1
cprio=0
for line in open(consofullpath):
    c= line.count(";")  
    if n>1 and c != cprio:
        f.write (line)
        split_lines=split_lines+1
    n=n+1
    cprio=c
print("Processed lines:"+ str(n-1))
print("Possible split/messed up lines:"+ str (split_lines))
```
