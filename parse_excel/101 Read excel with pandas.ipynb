```python
df_duplicates = df[df.duplicated(['Auth_ID'],keep=False)]
df_duplicates
```

```python
df_duplicates.to_excel("output/101_Duplicates.xlsx", sheet_name='Duplicates')
```

```python
df_deduped_ID= df['Auth_ID'].drop_duplicates()
```

```python
gap_list=[]
n= 0
for row in df_deduped_ID:
    if n==0:
        n= row-1
    
    if row-n > 1:
        gap_list.extend(range(n+1,row))
    n=row

df2 = pd.DataFrame({'Auth_ID':gap_list})
df2.to_excel("output/101_Auth_ID_Gaps.xlsx")
print(len(gap_list))
```
